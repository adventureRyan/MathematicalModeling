```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'SimSun'  # 使用宋体字体作为示例
```

```python
# 读取Excel文件
xl = pd.ExcelFile('2d数据.xlsx')
sheet_names = xl.sheet_names  # 获取所有表单名称
```

```python
def polyfit_plot(x, y, degree=2, **kwargs):
    """绘制多项式拟合曲线的函数"""
    coefficients = np.polyfit(x, y, degree)
    polynomial = np.poly1d(coefficients)
    x_sorted = np.linspace(min(x), max(x), 500)  # 生成x的密集值以获得更平滑的曲线
    y_poly = polynomial(x_sorted)
    plt.plot(x_sorted, y_poly, **kwargs)

for sheet in sheet_names:
    df = xl.parse(sheet)  # 读取每一个表单
    
    # 求描述性统计值
    blood_stats = df['血肿体积'].describe()
    water_stats = df['水肿体积'].describe()

    print(f"{sheet} 血肿体积的统计值:")
    print(blood_stats)
    print(f"{sheet} 水肿体积的统计值:")
    print(water_stats)
    
    # 绘制散点图和模拟曲线
    plt.figure(figsize=(12, 5))
    
    # 血肿体积随时间的变化
    plt.subplot(1, 2, 1)
    plt.scatter(df['时间'], df['血肿体积'], label='实际数据点', color='blue')
    polyfit_plot(df['时间'], df['血肿体积'], degree=2, color='red', label='2阶多项式拟合')
    plt.title(f"{sheet} - 血肿体积随时间的变化")
    plt.legend()
    
    # 水肿体积随时间的变化
    plt.subplot(1, 2, 2)
    plt.scatter(df['时间'], df['水肿体积'], label='实际数据点', color='blue')
    polyfit_plot(df['时间'], df['水肿体积'], degree=2, color='red', label='2阶多项式拟合')
    plt.title(f"{sheet} - 水肿体积随时间的变化")
    plt.legend()
    
    plt.tight_layout()
    plt.show()
```

```python
# 以某种方法如均值为例
means = []
for sheet in sheet_names:
    df = xl.parse(sheet)
    blood_mean = df['血肿体积'].mean()
    water_mean = df['水肿体积'].mean()
    means.append((blood_mean, water_mean))

# 转换为DataFrame
means_df = pd.DataFrame(means, columns=['血肿体积均值', '水肿体积均值'], index=sheet_names)

# 绘制条形图
means_df.plot(kind='bar', figsize=(15, 7))
plt.title("不同治疗方法的血肿体积和水肿体积均值")
plt.ylabel("体积")
plt.xlabel("治疗方法")
plt.tight_layout()
plt.show()
```

```python
stats_list = []

for sheet in sheet_names:
    df = xl.parse(sheet)  # 读取每一个表单
    
    # 计算描述性统计值
    blood_mean = df['血肿体积'].mean()
    blood_median = df['血肿体积'].median()
    blood_std = df['血肿体积'].std()

    water_mean = df['水肿体积'].mean()
    water_median = df['水肿体积'].median()
    water_std = df['水肿体积'].std()

    stats_list.append((blood_mean, blood_median, blood_std, water_mean, water_median, water_std))

# 创建一个新的DataFrame存储描述性统计值
columns = [
    '血肿体积_均值', '血肿体积_中位数', '血肿体积_标准偏差',
    '水肿体积_均值', '水肿体积_中位数', '水肿体积_标准偏差'
]
stats_df = pd.DataFrame(stats_list, columns=columns, index=sheet_names)

# 绘制均值图
stats_df[['血肿体积_均值', '水肿体积_均值']].plot(kind='bar', figsize=(15, 6))
plt.title("各种治疗方法的均值")
plt.ylabel("体积")
plt.xlabel("治疗方法")
plt.tight_layout()
plt.show()

# 绘制中位数图
stats_df[['血肿体积_中位数', '水肿体积_中位数']].plot(kind='bar', figsize=(15, 6))
plt.title("各种治疗方法的中位数")
plt.ylabel("体积")
plt.xlabel("治疗方法")
plt.tight_layout()
plt.show()

# 绘制标准偏差图
stats_df[['血肿体积_标准偏差', '水肿体积_标准偏差']].plot(kind='bar', figsize=(15, 6))
plt.title("各种治疗方法的标准偏差")
plt.ylabel("体积")
plt.xlabel("治疗方法")
plt.tight_layout()
plt.show()
```

# 相关性系数

```python
# 读取Excel文件
xl = pd.ExcelFile('2d数据.xlsx')
sheet_names = xl.sheet_names  # 获取所有表单名称

correlation_list = []

for sheet in sheet_names:
    df = xl.parse(sheet)  # 读取每一个表单
    
    # 计算水肿体积与血肿体积之间的相关系数
    correlation = df[['水肿体积', '血肿体积']].corr().iloc[0, 1]
    
    correlation_list.append(correlation)

# 创建一个新的DataFrame存储相关系数
correlation_df = pd.DataFrame(correlation_list, columns=['相关系数'], index=sheet_names)
print(correlation_df)

# 绘制柱状图
correlation_df.plot(kind='bar', figsize=(15, 6), color='skyblue')
plt.title("各种治疗方法中水肿体积与血肿体积的相关系数")
plt.ylabel("相关系数")
plt.xlabel("治疗方法")
plt.tight_layout()
plt.axhline(0, color='black',linewidth=0.5)  # 添加y=0的水平线
plt.show()
```

# 方差分析

```python
import pandas as pd
from scipy.stats import f_oneway
```

```python
xl = pd.ExcelFile('2d数据.xlsx')
sheet_names = xl.sheet_names  # 获取所有表单名称

# 对于血肿体积和水肿体积分别提取所有治疗方法的数据
blood_volumes = [xl.parse(sheet)['血肿体积'].values for sheet in sheet_names]
water_volumes = [xl.parse(sheet)['水肿体积'].values for sheet in sheet_names]
```

```python
# 对血肿体积进行ANOVA
f_stat_blood, p_value_blood = f_oneway(*blood_volumes)
print(f"血肿体积：F-statistic: {f_stat_blood}, P-value: {p_value_blood}")

# 对水肿体积进行ANOVA
f_stat_water, p_value_water = f_oneway(*water_volumes)
print(f"水肿体积：F-statistic: {f_stat_water}, P-value: {p_value_water}")
```

解释结果：
F统计量（F-statistic）表示组间变异与组内变异之间的比率。值越大，表示不同组之间的差异越大。
P值（P-value）表示观察到的数据或更极端数据出现的概率，假设所有治疗方法在统计上是相同的。通常，如果P值小于0.05，我们会拒绝零假设，认为至少有两种治疗方法在统计上是不同的。

### 血肿体积：

F-statistic (血肿体积) = 2.6784: 这个值表示组间差异与组内差异之间的比率。F统计量的值越大，表示不同组之间的差异越大。在这种情况下，F统计量的值为2.6784，表示存在一定的组间差异。

P-value (血肿体积) = 0.0003588: P值远小于0.05（通常使用的显著性水平），这意味着我们有足够的证据拒绝零假设。因此，我们可以得出结论，不同的治疗方法对血肿体积有统计学上的显著差异。

### 水肿体积：

F-statistic (水肿体积) = 4.6325: 与血肿体积的F统计量相比，这个值更大，表示在不同治疗方法之间，对水肿体积的影响存在更大的差异。

P-value (水肿体积) = 6.4715e-09: 这是一个非常小的P值，远小于0.05。这意味着我们有足够的证据拒绝零假设。所以，我们可以确定地说，不同的治疗方法对水肿体积有非常显著的统计差异。

总结：
基于上述F统计量和P值，我们可以得出结论，不同的治疗方法对血肿体积和水肿体积都有统计学上的显著差异。特别是对于水肿体积，治疗方法间的差异更为显著。这可能意味着某些治疗方法对于减少或控制血肿和水肿体积可能更为有效。

# 回归分析

```python
import pandas as pd
import statsmodels.api as sm
```

```python
xl = pd.ExcelFile('2d数据.xlsx')
sheet_names = xl.sheet_names

dataframes = []
for idx, sheet in enumerate(sheet_names):
    df = xl.parse(sheet)
    df['治疗方法'] = idx  # 为每个治疗方法添加一个数字标识符
    dataframes.append(df)

all_data = pd.concat(dataframes, ignore_index=True)
```

```python
dummies = pd.get_dummies(all_data['治疗方法'], prefix='治疗方法', drop_first=True)
all_data = pd.concat([all_data, dummies], axis=1)
```

```python
X = all_data[['时间', '血肿体积'] + [col for col in dummies.columns]]
X = sm.add_constant(X)  # 添加一个常数列，即intercept
y = all_data['水肿体积']
```

```python
model = sm.OLS(y, X).fit()
print(model.summary())
```

解释结果：
模型摘要：这将给您一个详细的回归输出，包括每个自变量的系数、t值、P值和R-squared值等。

系数：这些描述了当其他所有变量保持不变时，每个自变量与因变量之间的关系。

P值：如果某个自变量的P值小于0.05，那么通常认为这个自变量对因变量有显著的影响。

R-squared：描述了模型中所有自变量解释的因变量变异的比例。


模型的总体统计指标：

R-squared (R²): 0.186，这是模型的决定系数，表示模型中独立变量对因变量的解释能力。这意味着模型中的变量可以解释18.6%的因变量（水肿体积）的变异。
Adjusted R-squared: 0.146，考虑了模型中独立变量的数量，通常比R²小。它为14.6%，说明模型可能包括了部分不重要的解释变量。
F-statistic: 4.619，表示模型的整体显著性。F统计量用于检验模型中所有独立变量的回归系数是否全部等于0。
Prob (F-statistic): 1.39e-09，这是F统计量的P值。这个非常小的P值意味着我们可以拒绝所有回归系数都为0的原假设，即模型是显著的。
AIC and BIC: 这两个指标是模型的信息准则，用于模型选择。较小的值意味着更好的模型。
变量的系数和统计显著性：

coef: 每个变量的系数。例如，血肿体积的系数为0.1699意味着，当其他变量保持不变时，血肿体积每增加1单位，预期的水肿体积增加约0.1699单位。
std err: 表示系数的标准误。
t: 每个系数的t统计量。
P>|t|: t统计量的P值，用于检验系数是否显著地不等于0。例如，血肿体积的P值为0.005，远小于0.05，因此它是统计显著的。
[0.025 0.975]: 表示回归系数的95%置信区间。
其他统计信息:

Omnibus and Prob(Omnibus): 这些统计量用于检验残差的正态性。Prob(Omnibus)的值为0，这意味着残差不是正态分布的。
Durbin-Watson: 用于检测残差的自相关。其值为1.060，接近1，可能表示存在残差的正自相关。
Jarque-Bera (JB) and Prob(JB): 另一种检验残差正态性的方法。再次确认残差不是正态分布的。
Skew: 表示残差的偏度。正值意味着残差分布是右偏的。
Kurtosis: 表示残差的峰度。
Cond. No.: 表示模型的条件数。这个大的值（1.86e+06）表明存在强的多重共线性或其他数值问题。
结论:

模型只解释了水肿体积18.6%的变异。
血肿体积与水肿体积之间存在统计显著的正相关。
其中治疗方法11和治疗方法12在0.05的显著性水平下与水肿体积显著相关。
存在某些模型假设不满足的迹象，例如残差的非正态性和自相关。此外，可能存在多重共线性问题。
建议进一步诊断模型，尤其是检查模型的假设，以及考虑进行特征选择或变量转换来优化模型。

```python
predicted = model.fittedvalues
sns.scatterplot(x=y, y=predicted)
plt.plot([min(y), max(y)], [min(y), max(y)], '--', color='red')
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.title('Actual vs Predicted Values')
plt.show()
```

### 系数的置信区间

```python
coefficients = model.params
confidence_intervals = model.conf_int()

plt.figure(figsize=(10, 6))
plt.errorbar(coefficients.index, coefficients.values, 
             yerr=[coefficients.values - confidence_intervals[0], 
                   confidence_intervals[1] - coefficients.values], 
             fmt='o', capsize=5)
plt.xticks(rotation=45)
plt.ylabel('Coefficient value')
plt.title('Regression Coefficients and Confidence Intervals')
plt.tight_layout()
plt.show()
```

# 交互效应分析

```python
import statsmodels.api as sm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. 读入和整合数据
xls = pd.ExcelFile('2d数据.xlsx')
sheet_names = xls.sheet_names

all_data = pd.DataFrame()

for idx, sheet in enumerate(sheet_names):
    df = xls.parse(sheet)
    df['治疗方法'] = idx + 1
    df = df.sort_values(by='时间')  # 这里进行排序
    all_data = all_data.append(df, ignore_index=True)

# 2. 模型设定和交互效应分析
df_dummies = pd.get_dummies(all_data['治疗方法'], prefix='治疗方法')

all_data_new = pd.concat([all_data, df_dummies], axis=1)

for column in df_dummies.columns:
    all_data_new[f"{column}_时间"] = all_data_new[column] * all_data_new['时间']

Y = all_data_new['水肿体积']
X = all_data_new[['时间'] + list(df_dummies.columns) + [f"{column}_时间" for column in df_dummies.columns]]
X = sm.add_constant(X)

model = sm.OLS(Y, X).fit()

# 输出摘要结果
print(model.summary())

# 3. 绘制交互效应图
plt.figure(figsize=(12, 6))

for idx, sheet in enumerate(sheet_names):
    treatment_data = all_data[all_data['治疗方法'] == idx + 1]
    plt.plot(treatment_data['时间'], treatment_data['水肿体积'], label=f'治疗方法{idx + 1}')

plt.xlabel('时间')
plt.ylabel('水肿体积')
plt.title('时间与水肿体积的交互效应图')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
